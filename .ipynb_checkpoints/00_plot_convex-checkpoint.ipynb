# Plot convex and non convex

```python
import numpy as np
import pandas as pd

import plotly.graph_objects as go
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

%matplotlib inline
```

```python
u, v = np.mgrid[-1:1:100j, -1:1:100j]
```

```python
z = u**2 + v**2
```

```python
fig = go.Figure(data=[go.Surface(z=z, x=u, y=v)])

fig.update_layout(title='$CONVEX: z = x^2 + y^2$', autosize=False,
                  width=600, height=600,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```

```python
z = u*v
```

```python
fig = go.Figure(data=[go.Surface(z=z, x=u, y=v)])

fig.update_layout(title="$NONCONVEX: z = x \times y$", autosize=False,
                  width=600, height=600,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```
